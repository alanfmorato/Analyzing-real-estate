```python
#Quantidade de apartamentos
apto = df['Tipo'] == 'Apartamento'
apto
freq_apto = df[apto].shape[0]
freq_apto
```

```python
#Quantidade de casas, casas de condominio e casas de vilas
casa = (df['Tipo'] == 'Casa') | (df['Tipo'] == 'Casa de Condomínio') | (df['Tipo'] == 'Casa de Vila')
casa
freq_casa = df[casa].shape[0]
freq_casa
```

```python
#Quantidade de casa com áreas entre 60 e 100 m^2
area = (df['Area'] >= 60) & (df['Area'] <= 100)
area
freq_area = df[area].shape[0]
freq_area
```

```python
#Quantidade de casas com pelo menos 4 quartos e valor menor que R$2000,00
condicao = (df['Quartos'] >= 4) & (df['Valor'] <= 2000)
condicao
freq_condicao = df[condicao].shape[0]
freq_condicao
```
